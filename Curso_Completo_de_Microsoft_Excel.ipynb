## 🟡 **Módulo 2: Fórmulas y Funciones**

### 2.1. Diferencia entre fórmula y función

* **Fórmula**: operación matemática escrita manualmente. Ejemplo: `=A1+B1`.
* **Función**: comando predefinido en Excel. Ejemplo: `=SUMA(A1:A5)`.

---

### 2.2. Funciones básicas

1. **SUMA** → `=SUMA(A1:A5)`
2. **PROMEDIO** → `=PROMEDIO(A1:A5)`
3. **MAX y MIN** → `=MAX(A1:A5)`, `=MIN(A1:A5)`
4. **CONTAR** → `=CONTAR(A1:A5)`

👉 **Ejemplo práctico**:
En una lista de calificaciones, calcula:

* Total de puntos.
* Promedio del grupo.
* La calificación más alta y más baja.

📸 **Imagen sugerida**: veace en el libro de excel

---

### 2.3. Funciones condicionales

1. **SI** → `=SI(B2>=6,"Aprobado","Reprobado")`
2. **Y** / **O** → combinaciones lógicas.

👉 **Ejemplo práctico**:
En una lista de alumnos, mostrar “Aprobado” si la nota ≥6, de lo contrario “Reprobado”.
